1. Using the mean squared error (MSE) as your objective function allows for a closed form solution to finding the maximum likelihood estimate (MLE) of your model parameters in linear regression. Let’s consider the simple, single predictor variable model $Y= \beta_0 + \beta_1 X $. 

a) Use algebra to show how you can expand out the MSE to get from i to ii below. Do your proof using Latex equation formatting.


 i) $MSE(\beta_0, \beta_1)=E[ (Y-(\beta_0 + \beta_1 X))^2]$ 

**Answer:** 

$MSE(\beta_0, \beta_1)$

>   $=E[ (Y-(\beta_0 + \beta_1 X))^2]$

>   $=E[Y^2] - \beta_0E[Y] - 2\beta_1 E[X,Y] + E[(\beta_0 + \beta_1 X)]$

>   $=E[Y^2] - \beta_0E[Y] - 2\beta_1 Cov[X,Y] -2 \beta_1 E[X]E[Y] + \beta_0^2 + 2 \beta_0 \beta_1 E[X] + \beta_1^2 Var[X] + \beta_1^2 (E[X])^2$


                  

b) Prove that the MLE of $\beta_0$ is $E[Y]- \beta_1 E[X]$ by taking the derivative of _ii_ above, with respect to $\beta_0$, setting the derivative to zero, and solving for $\beta_0$.

**Answer:**

$MLE(\beta_0)=E[Y] - \beta_1E[X]$

>   $f(\beta_0)=E[Y^2] - 2\beta_0E[Y] - 2\beta_1 Cov[X,Y] -2 \beta_1 E[X]E[Y] + \beta_0^2 + 2 \beta_0 \beta_1 E[X] + \beta_1^2 Var[X] + \beta_1^2 (E[X])^2$


>   $f'(\beta_0) = 0 - E[Y] - 0 - 0 + 2\beta_0 + 2\beta_1 E[X] + 0 + 0$

>   $ 0 = - 2E[Y] + 2\beta_0 + 2\beta_1 E[X]$

>   $2\beta_0 = 2E[Y] - 2\beta_1 E[X]$

>   $\beta_0 = E[Y] - \beta_1E[X]$



_ii)_ $=E[Y^2] -2 \beta_0E[Y]-2 \beta_1 Cov[X,Y]-2 \beta_1 E[X]E[Y]+ \beta_0^2 +2 \beta_0 \beta_1 E[X]+\beta_1^2 Var[X]+ \beta_1^2 (E[X])^2$

c) Prove that the MLE for $\beta_1$ is $Cov[X,Y]/Var[X]$ by taking the derivative of _ii_ above, with respect to $\beta_1$, setting the derivative to zero, and solving for $\beta_1$. 

**Answer:**

$MLE(\beta_1)= \frac{Cov[X,Y]}{Var[X]}$

>   $f(\beta_1)=E[Y^2] - 2\beta_0E[Y] - 2\beta_1 Cov[X,Y] -2 \beta_1 E[X]E[Y] + \beta_0^2 + 2 \beta_0 \beta_1 E[X] + \beta_1^2 Var[X] + \beta_1^2 (E[X])^2$

>   $f'(\beta_1)= 0 - 0 - 2Cov[X,Y] - 2E[X]E[Y] + 0 + 2\beta_0E[X] + 2\beta_1Var[X] + 2\beta_1(E[X])^2$

>   $0 = - 2Cov[X,Y] - 2E[X]E[Y] + 2\beta_0E[X] + 2\beta_1Var[X] + 2\beta_1(E[X])^2$ 

>   $- 2\beta_1Var[X] - 2\beta_1(E[X])^2 = - 2Cov[X,Y] - 2E[X]E[Y] + 2\beta_0E[X]$

>   $-2\beta_1(Var[X] + (E[X])^2) = -2(Cov[X,Y] + E[X]E[Y] - \beta_0E[X])$

>   $\beta_1(Var[X] + (E[X])^2) = Cov[X,Y] + E[X]E[Y] - \beta_0E[X]$

>   $\beta_1 = \frac{Cov[X,Y] + E[X]E[Y] - \beta_0E[X]}{Var[X] + (E[X])^2}$

3. In the case where X and Y are standard normal variables (i.e. $N(0,1)$), show that the ordinary least squares solution converges to the same solution as the Pearson correlation coefficient $r=Cov[X,Y]/(\sigma_X * \sigma_Y)$. What exactly do they both converge to?


**Answer:** 

The ordinary least squares solution can be expressed as $ \hat{\beta} = \frac{ \sum_{i=1}^{n} (x_i-\bar{x})(y_i - \bar{y})} {\sum_{i=1}^{n} (x_i - \bar{x})^2} = \frac{Cov[X,Y]}{Var[X]} $. The Pearson correlation coefficient can be expressed as $r = \frac{Cov[XY]}{\sigma_X \sigma_Y}$, and when $X$ and $Y$ are both standard normal distributions $ (\sigma_X \sigma_Y)= 1$, and $ Var[X]=1 $, which allows both the ordinary least squares solution and the Pearson correlation coefficient to converge to $ Cov[X,Y] $.

4. Three methods for evaluating a regression model’s performance are $r^2$, $RSE$, and the $F$ Test. Show the functional form of each method and describe conceptually what it is telling you about your model.

**Answer:**

The functional form of the $r^2$ is $ R^2 = 1 - \frac{\sum_{i} (y_i - f_i)^2 } {\sum_{i} (Y_i - \bar Y)^2}$ and it tells us how well the model fits the data, known as the "goodness of fit." The numerator in the fraction is also known as the residual sum of squares and the denominator is also known as the total sum of squares.

The functional form of the RSE is $ RSE = \sigma_{model}^2 = \sqrt{ \frac{RSS}{n-2} } $ and it tells us how certain we can be on the residual error value of our model.

The functional form of the $F$ Test is 
$ F = \frac{\sigma_{between}^2}{\sigma_{within}^2} = \frac{ \sum_{i=1}^{k} \frac{n_i(y_i - \bar{y})^2}{K-1}} {\frac{ \sum_{i=1}^{k} \sum_{j=1}^{n_i} (y_{i,j} - \bar{y}_i)^2 } {N-K} }  $ and it explains the proportion of explained variance over unexplained variance or between-group variability over within-group variability.

7. A linear mixed effect model has the form $Y = X\beta + Z\upsilon + \epsilon$. 

a) What are the assumptions about $Z$ and $\upsilon$ that make this different from using a regular control variable in a simple linear regression model?


**Answer:** One assumption is that $vZ$ is random and that a unit change in $Z$ does not necessarily cause an up or down effect on Y. Furthermore, Z is assumed to have its own "random structure" which reflects that the variance of the regression coefficient for the random effects is different that the variance for the regression coefficients for the fixed effects.

b) What is the objective function for this model (i.e., the function that needs to be minimized to find the optimal parameters) and how is it different than the objective function for the ordinary least squares model?


**Answer:**

The objective function that needs to be minimized is $ min(||Y-X\beta-Z\Lambda_{\theta}\upsilon||^2+||\upsilon||^2) $ and it is different from the objective function for the ordinary least squares model by accounting for the variance of the random effects that explains the random structure in $Z$, and it takes into account the different variance of the regression coefficients on the random effects.

c) What advantage does including a random effects term in your model give you with regards to understanding the fixed effects relationship?

**Answer:** An advantage of including a random effects term is that it allows you to "control" for known sources of variability or factors that can confound your analysis. Furthermore, this would allow you to have a better understanding of the true impact the fixed effects have on determining Y.

8. Load the cbpp dataset from the lme4 library. Run ?cbpp after loading the lme4 library to get information on the dataset.

(c) Use the lmer function to include the herd identifier as a random effect. How does this impact the fixed effect of herd size on CBPP incidence?

(d) Compare the simple linear model (i.e., the fixed effects only model) with the mixed effect model using AIC (https://en.wikipedia.org/wiki/Akaike_information_criterion). Does adding the random effect to the model improve or change the fixed effect?


```r
# -------------------------------
# Question 8
# -------------------------------


# (a)

library(lme4)
library(ggplot2)

ggplot(cbpp, aes(x=size,y=incidence)) + geom_point()
```

```r
# (b)
lm.fit = lm(incidence~size, data=cbpp)

summary(lm.fit)
coef(lm.fit)

# There is a positive relationship between herd size and the rate of incidence of disease.
# Herd size impacts disease rate by a factor of approximately 0.138 (Std. Error = 0.0439) 
# for every unit increase in herd size.
```

1. How does knowing data type (e.g., qualitative vs. quantitative) provide insights into the type of distribution a random variable comes from?

**Answer:** Knowing the type of the data will allow you to know the type of distribution the variable comes from, because of the values the variable can and cannot take on. Quantitative data is continuous so the random variable would come from a continuous distribution, whereas qualitative data is a categorical measure so the random variable would come from a discrete distribution.

3. Using the *runif* function, run three experiments by simulating the outcomes of rolling a single, six-sided die. Show the distribution of each experiment. Show how the simulated means compare to the expected mean of a fair roll.
        	a) Exp 1: 10 independent throws
        	b) Exp 2: 1,000 independent throws
        	c) Exp 3: 10,000 independent throws


**Answer:** The simulated means are plotted in red and the expected mean of a fair roll is plotted in green in the histograms below. In summary, the more throws measured increases the likelihood that the two means are equal.

```r
# -------------------------------
# Question 3
# -------------------------------
library(ggplot2)
require(gridExtra)


#Exp 1: 10 independent throws
data1 <- round(runif(10, min =1 , max = 6), 0) #Round values to the nearest digit
exp1 <- data.frame(data1) #Make a data frame of values
plot1 <- ggplot(exp1, aes(x=exp1)) + geom_histogram(binwidth=1, color="gray", fill="white") + #Plots histogram
            geom_vline(aes(xintercept=mean(data1)), color="red", linetype="dashed", size=1) + #Observed mean
            geom_vline(aes(xintercept=mean(c(1,2,3,4,5,6))), color="green", linetype="dashed", size=1) #Expected mean

#Exp 2: 1,000 independent throws
data2 <- round(runif(1000, min =1 , max = 6), 0)
exp2 <- data.frame(data2)
plot2 <- ggplot(exp2, aes(x=exp2)) + geom_histogram(binwidth=1, color="gray", fill="white") +
            geom_vline(aes(xintercept=mean(data2)), color="red", linetype="dashed", size=1) +
            geom_vline(aes(xintercept=mean(c(1,2,3,4,5,6))), color="green", linetype="dashed", size=1) #Expected mean

#Exp 3: 10,000 independent throws
data3 <- round(runif(10000, min =1 , max = 6), 0)
exp3 <- data.frame(data3)
plot3 <- ggplot(exp3, aes(x=exp3)) + geom_histogram(binwidth=1, color="gray", fill="white") +
            geom_vline(aes(xintercept=mean(data3)), color="red", linetype="dashed", size=1) +
            geom_vline(aes(xintercept=mean(c(1,2,3,4,5,6))), color="green", linetype="dashed", size=1) #Expected mean

mean(data1)
mean(data2)
mean(data3)
grid.arrange(plot1, plot2, plot3, ncol=3)
```

4. Instead of rolling one die, run a set of experiments reporting the outcomes of rolling multiple dice at the same time.
        	a) Exp 1: 10,000 throws, 1 die
        	b) Exp 2: 10,000 throws, 2 dice
        	c) Exp 3: 10,000 throws, 3 dice
        	d) Exp 4; 10,000 throws, 6 dice.

Show the distribution of results for each experiment. Which of these experiments produces a distribution most similar to a normal normal distribution? Justify your conclusion using Q-Q plots.

**Answer:** The last one produces the most similar to a normal distribution, because as number of samples from the possible dice roll increases the relationship with the theoretical normal distribution becomes more linear as seen in the Q-Q plots.

```r
# -------------------------------
# Question 4
# -------------------------------


#DISTRIBUTION OF RESULTS

library(ggplot2)
require(gridExtra)

#Exp 1: 10,000 independent throws for 1 die
dicex1 <- c(1:6)
data1 <- round(runif(10000, min =1 , max = 6), 0) #Round values to the nearest digit
exp1 <- data.frame(data1) #Make a data frame of values
plot1 <- ggplot(exp1, aes(x=exp1)) + geom_histogram(binwidth=1, color="gray", fill="white") + #Plots histogram
            geom_vline(aes(xintercept=mean(data1)), color="red", linetype="dashed", size=1) + #Observed mean
            geom_vline(aes(xintercept=mean(dicex1)), color="green", linetype="dashed", size=1) #Expected mean

#Exp 2: 1,000 independent throws for 2 dice
dicex2 <- c(2:12)
data2 <- round(runif(10000, min =2 , max = 12), 0)
exp2 <- data.frame(data2)
plot2 <- ggplot(exp2, aes(x=exp2)) + geom_histogram(binwidth=1, color="gray", fill="white") +
            geom_vline(aes(xintercept=mean(data2)), color="red", linetype="dashed", size=1) +
            geom_vline(aes(xintercept=mean(dicex2)), color="green", linetype="dashed", size=1) #Expected mean

#Exp 3: 10,000 independent throws for 3 dice
dicex3 <- c(3:18)
data3 <- round(runif(10000, min =3 , max = 18), 0)
exp3 <- data.frame(data3)
plot3 <- ggplot(exp3, aes(x=exp3)) + geom_histogram(binwidth=1, color="gray", fill="white") +
            geom_vline(aes(xintercept=mean(data3)), color="red", linetype="dashed", size=1) +
            geom_vline(aes(xintercept=mean(dicex3)), color="green", linetype="dashed", size=1) #Expected mean

#Exp 4: 10,000 independent throws for 6 dice
dicex4 <- c(6:36)
data4 <- round(runif(10000, min =6 , max = 36), 0)
exp4 <- data.frame(data4)
plot4 <- ggplot(exp4, aes(x=exp4)) + geom_histogram(binwidth=1, color="gray", fill="white") +
            geom_vline(aes(xintercept=mean(data4)), color="red", linetype="dashed", size=1) +
            geom_vline(aes(xintercept=mean(dicex4)), color="green", linetype="dashed", size=1) #Expected mean

grid.arrange(plot1, plot2, plot3, plot4, ncol=2)
```

```r
qqplot1 <- ggplot(exp1, aes(sample=exp1)) + 
            stat_qq(distribution = qnorm) + ggtitle("Experiment 1")

qqplot2 <- ggplot(exp2, aes(sample=exp2)) + 
            stat_qq(distribution = qnorm) + ggtitle("Experiment 2")

qqplot3 <- ggplot(exp3, aes(sample=exp3)) + 
            stat_qq(distribution = qnorm) + ggtitle("Experiment 3")

qqplot4 <- ggplot(exp4, aes(sample=exp4)) + 
            stat_qq(distribution = qnorm) + ggtitle("Experiment 4")

grid.arrange(qqplot1, qqplot2, qqplot3, qqplot4, ncol=2)
```

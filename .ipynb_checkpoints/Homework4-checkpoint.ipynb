1. Consider the problem of classifying a binary response variable (i.e., $y \in \{1,0\}$). If there is no overlap in the values of X when y = 1 and when y=0,  such that there is a large “gap” between the two distributions of X values, then this is problematic for one of the classifiers discussed in class and the text. What classifier does this situation pose a problem for? Explain conceptually why this is a problem and compare it with another classifier approach that does not suffer this limitation.

**Answer:** This poses a problem to logistic regression models, which can be explained as not being able to narrow down the possible curves the model can take on in order to explain the difference between the two distributions. In logistic regression, the curve could take on many different values which makes it difficult to determine the optimal model, unlike logistic regression kNN classification would not be affected by this large "gap" because of its use of finding the nearest neighbors for classifications.

2. Compare logistic regression, LDA, and kNN classification approaches. Which are parametric which are non-parametric? For parametric models what functions do they assume? For non-parametric methods, how do the classifiers separate groups? How is the flexibility/bias tradeoff adjusted for each method?




**Answer:** Logistic regression and LDA are both parametric classification methods, however kNN is a non-parametric classification method. Parametric models assume that the data used follows a specific distribution, whereas non-parametric approaches do not. The kNN of classification or forming decision boundaries is based on using the distance of the nearest neighbors, which requires no assumptions based on the distribution of the data. For the kNN classification method the flexibility/bias tradeoff is adjusted by choosing the value of k. LDA is more biased than it is flexible due to its method of estimating parameters.

3. What is the curse of dimensionality? Why is it especially problematic for kNN classification (i.e., why does kNN fail in high dimensional contexts)?


**Answer:** The curse of dimensionality is essentially overfitting, because at high dimensions the value added by the additional dimensions is much smaller compared to the complexity it adds to the model. As dimensionality increases, the absolute error of the kNN regression model increases, and when dimensionality gets too high kNN regression completely fails. This occurs, because sparsity increases and the method is not able to find close enough neighbors in order to make an accurate classification.

Based on these results, which method should we prefer to use for classification of new observations? Why?

**Answer:** Since kNN uses the response itself during training, its error rate is always 0% on training data, which means that the error rate on the test data was 36%. The prefered method would have the lowest test data error rate which is the logistic regression model.

* For KNN with K=1, the training error rate is 0% because for any training observation, its nearest neighbor will be the response itself. So, that means that KNN has a test error rate of 36%. Thus logistic regression is preferred because of its lower test error rate of only 30%.

5. This question should be answered using the Weekly data set, which is part of the ISLR package. This data is similar in nature to the Smarket data from this chapter’s lab, except that it contains 1,089 weekly returns for 21 years, from the beginning of 1990 to the end of 2010.

(a) Produce some numerical and graphical summaries of the Weekly data. Do there appear to be any patterns?

(b) Use the full data set to perform a logistic regression with Direction as the response and the five lag variables plus Volume as predictors. Use the summary function to print the results. Do any of the predictors appear to be statistically significant? If so, which ones?

(c) Compute the confusion matrix and overall fraction of correct predictions. Explain what the confusion matrix is telling you about the types of mistakes made by logistic regression.

(d) Now fit the logistic regression model using a training data period from 1990 to 2008, with Lag2 as the only predictor. Compute the confusion matrix and the overall fraction of correct predictions for the held out data (that is, the data from 2009 and 2010).

(e) Repeat (d) using LDA.

(f) Repeat (d) using QDA.

(g) Repeat (d) using KNN with K = 1.

(h) Which of these methods appears to provide the best results on
this data?

(i) Experiment with different combinations of predictors, includ- ing possible transformations and interactions, for each of the methods. Report the variables, method, and associated confu- sion matrix that appears to provide the best results on the held out data. Note that you should also experiment with values for K in the KNN classifier.

```r
# ------------------
# Exercize 5
# ------------------

# (a).

require(ISLR)

data(Weekly)
summary(Weekly)

# Yes there appears to be patterns the description values for all of the Lag variables are highly similar if not the same.
```

```r
# (b).

glm.fits = glm(Direction ~ Lag1 + Lag2 + Lag3 + Lag4 + Lag5 + Volume, data=Weekly, family="binomial")
summary(glm.fit)

# Lag2 and the intercept appear to be significant
```

```r
# (c).

glm.probability = predict(glm.fits, type="response")
glm.predictions = ifelse(glm.probability > 0.5, "Up", "Down")
confusionMatrix = table(Weekly$Direction, glm.predictions)
confusionMatrix

# The diagonal values are correct predictions and the off-diagonals are the incorrect predictiosn. 
# Our model incorrectly predicted Up when it was actually Down a great deal. 
# Whereas there was a much smaller error rate for predicting Down when it was actually Up.
```

```r
# (d).

trainingSet = (Weekly$Year <= 2008)
testSet = Weekly[!trainingSet,]

glm.fit.Lag2 = glm(Direction ~ Lag2, data=Weekly, subset=trainingSet,
                  family="binomial")
glm.probability.Lag2 = predict(glm.fit.Lag2, type="response", newdata=testSet)
glm.predictions.Lag2 = ifelse(glm.probability.Lag2 > 0.5, "Up", "Down")
confusionMatrix.Lag2 = table(testSet$Direction, glm.predictions.Lag2)
confusionMatrix.Lag2

fraction.glm = (confusionMatrix.Lag2["Down", "Down"] +
            confusionMatrix.Lag2["Up", "Up"])/
            sum(confusionMatrix.Lag2)
fraction.glm
```

```r
# (e).
library(MASS)

lda.fit.Lag2 = lda(Direction ~ Lag2, data=Weekly, subset=trainingSet)
lda.predictions.Lag2 = predict(lda.fit.Lag2, newdata=testSet)
confusionMatrix.Lag2 = table(testSet$Direction, lda.predictions.Lag2$class)
confusionMatrix.Lag2

fraction.lda = (confusionMatrix.Lag2["Down", "Down"] +
            confusionMatrix.Lag2["Up", "Up"])/
            sum(confusionMatrix.Lag2)
fraction.lda
```

```r
# (f).

qda.fit.Lag2 = qda(Direction ~ Lag2, data=Weekly, subset=trainingSet,
                  family="binomial")
qda.predictions.Lag2 = predict(qda.fit.Lag2, newdata=testSet)
confusionMatrix.Lag2 = table(testSet$Direction, qda.predictions.Lag2$class)
confusionMatrix.Lag2

fraction.qda = (confusionMatrix.Lag2["Down", "Down"] +
            confusionMatrix.Lag2["Up", "Up"])/
            sum(confusionMatrix.Lag2)
fraction.qda
```

```r
# (g).
library(class)
set.seed(1)


trainingSetkNN = Weekly[trainingSet, c("Lag2", "Direction")]
knn.predictions.Lag2 = knn(train=data.frame(trainingSetkNN$Lag2), 
                           test=data.frame(testSet$Lag2),
                           cl=trainingSetkNN$Direction, k=1)
confusionMatrix.Lag2 = table(testSet$Direction, knn.predictions.Lag2)
confusionMatrix.Lag2

fraction.knn = (confusionMatrix.Lag2["Down", "Down"] +
            confusionMatrix.Lag2["Up", "Up"])/
            sum(confusionMatrix.Lag2)
fraction.knn
```

```r
# (h).

rbind(fraction.glm, fraction.lda, fraction.qda, fraction.knn)

# Glm and LDA
```

6. In this problem, you will develop a model to predict whether a given car gets high or low gas mileage based on the Auto data set.

(a) Create a binary variable, mpg01, that contains a 1 if mpg contains a value above its median, and a 0 if mpg contains a value below its median. You can compute the median using the median() function. Note you may find it helpful to use the data.frame() function to create a single data set containing both mpg01 and the other Auto variables.

(b) Explore the data graphically in order to investigate the associ- ation between mpg01 and the other features. Which of the other features seem most likely to be useful in predicting mpg01? Scatterplots and boxplots may be useful tools to answer this ques- tion. Describe your findings.

(c) Split the data into a training set and a test set.

(d) Perform LDA on the training data in order to predict mpg01 using the variables that seemed most associated with mpg01 in (b). What is the test error of the model obtained?

(e) Perform QDA on the training data in order to predict mpg01 using the variables that seemed most associated with mpg01 in (b). What is the test error of the model obtained?

(f) Perform logistic regression on the training data in order to pre- dict mpg01 using the variables that seemed most associated with mpg01 in (b). What is the test error of the model obtained?

(g) Perform KNN on the training data, with several values of K, in order to predict mpg01. Use only the variables that seemed most associated with mpg01 in (b). What test errors do you obtain? Which value of K seems to perform the best on this data set?

```r
# ------------------
# Exercize 6
# ------------------

# (a.)

data(Auto)
Auto$mpg01 = with(ifelse(mpg > median(mpg), "1", "0"), data=Auto)
```

```r
# (b.)

plot(Auto)
# difficult to assess with scatterplots because of binary nature
```

```r
# (c).

set.seed(1)
trainingSet = Auto[sample(x=nrow(Auto), size=0.5*nrow(Auto)),]
testSet = Auto[-sample(x=nrow(Auto), size=0.5*nrow(Auto)),]
```

```r
# (d).
library(MASS)

lda.fit = lda(mpg01 ~ displacement + horsepower + weight + acceleration + year + cylinders + origin, data=trainingSet)
lda.pred = predict(lda.fit, testSet)
table(testSet$mpg01, lda.pred$class)
sum(lda.pred$class != testSet$mpg01)/nrow(testSet)*100
```

```r
# (e).

qda.fit = qda(mpg01 ~ displacement + horsepower + weight + acceleration + year + cylinders + origin, data=trainingSet)
qda.pred = predict(qda.fit, testSet)
table(testSet$mpg01, qda.pred$class)
sum(qda.pred$class != testSet$mpg01)/nrow(testSet)*100
```

```r
# (f).

glm.fit = glm(as.factor(mpg01) ~ displacement + horsepower + weight + acceleration + year + cylinders + origin, data=trainingSet, family="binomial")
glm.probability = predict(glm.fit, testSet, type="response")
glm.prediction = ifelse(glm.probability > 0.5, "1", "0")
table(testSet$mpg01, glm.prediction)

sum(glm.prediction != testSet$mpg01)/nrow(testSet)*100
```

```r
# (g).

knn1.predictions = knn(train=data.frame(trainingSet$displacement), 
                           test=data.frame(testSet$displacement),
                           cl=trainingSet$mpg01, k=1)
confusionMatrix.knn1 = table(testSet$Direction, knn1.predictions)
confusionMatrix.knn1

sum(glm.prediction != testSet$mpg01)/nrow(testSet)*100
```
